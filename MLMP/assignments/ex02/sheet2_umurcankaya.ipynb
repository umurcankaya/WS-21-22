## 2 Linear regression with scikit-learn
Here we use scikit-learn to solve a linear regression problem
<a id='scikit-learn'></a>

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import ShuffleSplit
````

Let's first create a dataset:

````python
def f(x, m, b):
    return m * x + b
````

````python
npoints = 100
dmax = 100

x = np.linspace(0, 99, npoints).reshape(-1, 1) # x needs to be a 2D array
m = 2.0
b = 5.0
y_data = f(x, m, b).flatten() + np.random.randn(npoints) * dmax 
y_orig = f(x, m, b).flatten()
````

Let's first plot the data and have a look

````python
rs = ShuffleSplit(n_splits=1, test_size=.1, random_state=0)

for train, test in rs.split(x):
    print(f"TRAIN: {train}")
    print(f"TEST: {test}")
    
    plt.plot(x[:,0], y_orig, c='red', label='orig')
    plt.plot(x[:,0], y_data, '.', label = 'data')
    plt.plot(x[train],y_data[train], 'x', label = 'train')
    plt.plot(x[test],y_data[test], 'x', label = 'test')
    plt.legend(fontsize=15)
    plt.xlabel('x', fontsize=15)
    plt.ylabel('y', fontsize=15)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.show()
````

````python
# fit the data with a linear regression model:

reg = linear_model.LinearRegression()
reg.fit(x[train],y_data[train])   
m_fit = reg.coef_[0]         
b_fit = reg.intercept_         

print(f'Fitted slope = {m_fit:5.3f}')
print(f'Fitted intercept = {b_fit:5.3f}')
````

Compute y values predicted by the model and plot

````python
y_pred_all = reg.predict(x)         
y_pred_test = reg.predict(x[test])
y_pred_train = reg.predict(x[train])
````

````python
plt.plot(x,y_pred_all, c='red', label='pred', lw=3)
plt.plot(x[train],y_data[train],'x', c='orange', label = 'train')
plt.plot(x[test],y_data[test], 'x', label = 'test')
plt.legend(fontsize=15)
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
#plt.savefig('linear_regression-01.pdf', bbox_inches='tight')
plt.show()
````

````python
# get coefficient of determination R^2 

r2 = r2_score(y_data[test],y_pred_test)

print(f'R2 score = {r2:8.5f}')
````

````python
# get mean squared error and mean absolute error of test and training set

mse_test = mean_squared_error(y_data[test],y_pred_test)
mse_train = mean_squared_error(y_data[train],y_pred_train)

mae_test = mean_absolute_error(y_data[test],y_pred_test)
mae_train = mean_absolute_error(y_data[train],y_pred_train)

print(f'MSE test  = {mse_test:8.5f}    MAE test  = {mae_test:8.5f}')
print(f'MSE train = {mse_train:8.5f}     MAE train = {mae_train:8.5f}')
````

### 2.1 Random numbers

```np.random.randn(npoints)``` generates ```npoints``` points that are distributed according to standard normal distribution. It is used to disturb the artificially generated data with random noise.

```sklearn.model_selection.ShuffleSplit``` generates a random permutation of indices and splits them into two as training and test indices with their size determined by ```test_size``` argument. In our workflow it is used to shuffle and then to split the data.

### 2.2 Model

$f(x) = mx + b$

$m = $ ```reg.coef_[0]```

$b = $ ```reg.intercept_```

We used ```reg.predict``` method to make predictions with the fitted model.

### 2.3  Coefficient of determination

$$ R^2 = 1 - \frac{\sum_i (y_i - f(x_i))^2}{\sum_j (y_j - \bar{y})^2} $$
where $\bar{y} = \frac{1}{N}\sum_i y_i$ is the mean of the observed data $\vec{y}$. $R^2=1$ means that the predictions are exactly equal to the observations. $R^2=0$ means that our prediction is not better than using the observed mean as a prediction at every point, which indicates a bad fit. 

### 2.4 Dataset

````python
dmax_list = np.array([1,10,20])

for dmax in dmax_list:
    print('\ndmax = %d'%dmax)

    y_data = f(x, m, b).flatten() + np.random.randn(npoints) * dmax
    
    for train, test in rs.split(x):
        reg = linear_model.LinearRegression()
        reg.fit(x[train],y_data[train])  
    
    rs = ShuffleSplit(n_splits=1, test_size=.1, random_state=0)
    
    y_pred_all = reg.predict(x)         
    y_pred_test = reg.predict(x[test])
    y_pred_train = reg.predict(x[train])
    
    r2 = r2_score(y_data[test],y_pred_test)
    
    print(f'R2 score = {r2:8.5f}')
    
    mse_test = mean_squared_error(y_data[test],y_pred_test)
    mse_train = mean_squared_error(y_data[train],y_pred_train)

    mae_test = mean_absolute_error(y_data[test],y_pred_test)
    mae_train = mean_absolute_error(y_data[train],y_pred_train)

    print(f'MSE test  = {mse_test:8.5f}    MAE test  = {mae_test:8.5f}')
    print(f'MSE train = {mse_train:8.5f}     MAE train = {mae_train:8.5f}')    
````

MSE, MAE increase, and $R^2$ decrease with increasing $d_{max}$. This is an expected behaviour since increasing the error on the data makes it harder to infer the true relation.

## 3 Diabetes dataset
### 3.1 Fitting

````python
from sklearn import datasets

diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
````

````python
xcol=['age','sex','body mass index','average blood pressure','total serum cholesterol','low-density lipoproteins'
      ,'high-density lipoproteins','total cholesterol/HDL','serum triglycerides level','blood surgar level']

rs = ShuffleSplit(n_splits=1, test_size=.2, random_state=0)

for train_idx, test_idx in rs.split(diabetes_y):
    x_train, x_test = diabetes_X[train_idx].T, diabetes_X[test_idx].T
    y_train, y_test = diabetes_y[train_idx], diabetes_y[test_idx]
````

````python
fig, ax = plt.subplots(2,5, figsize=(15,6), sharey=True, sharex=True)
plt.subplots_adjust(wspace=0.05, hspace=0.3)

for col, axis in enumerate(ax.flatten()):
    axis.scatter(x_train[col], y_train, s=5, label='train')
    axis.scatter(x_test[col], y_test, s=5, label='test')
    
    axis.set(xlabel=xcol[col])
    
ax[0,0].legend(loc=(0.02,1.02), ncol=2, scatterpoints=3);
ax[0,0].set_ylabel('disease progression')
ax[1,0].set_ylabel('disease progression')

plt.tight_layout()
````

````python
reg_list = [linear_model.LinearRegression().fit(x_train[col][:,np.newaxis],y_train) for col in range(10)]
````

````python
x_pred = np.linspace(np.min(x_train),np.max(x_train),100)
y_pred = [reg.predict(x_pred[:,np.newaxis]) for reg in reg_list]
````

````python
fig, ax = plt.subplots(2,5, figsize=(15,6), sharey=True, sharex=True)
plt.subplots_adjust(wspace=0.05, hspace=0.3)

for col, axis in enumerate(ax.flatten()):
    axis.scatter(x_train[col], y_train, s=5, label='train', zorder=2)
    axis.scatter(x_test[col], y_test, s=5, label='test', zorder=2)
    axis.plot(x_pred, y_pred[col], lw=1.25, c='k', label='prediction', zorder=1)
    
    axis.set(xlabel=xcol[col])
    
ax[0,0].legend(loc=(0.02,1.02), ncol=2, scatterpoints=3);
ax[0,0].set_ylabel('disease progression')
ax[1,0].set_ylabel('disease progression')

plt.tight_layout()
````

````python
R2_list, mse_test, mse_train, mae_test, mae_train = np.zeros((5,10)) 

for col in range(10):
    
    y_pred_test = reg_list[col].predict(x_test[col][:,np.newaxis])
    y_pred_train = reg_list[col].predict(x_train[col][:,np.newaxis])
    
    R2_list[col] = r2_score(y_test, y_pred_test)

    mse_test[col], mse_train[col] = mean_squared_error(y_test,y_pred_test), mean_squared_error(y_train,y_pred_train)
    mae_test[col], mae_train[col] = mean_absolute_error(y_test,y_pred_test), mean_absolute_error(y_train,y_pred_train)
````

````python
fig, ax = plt.subplots(3,1, figsize=(6,6), sharex=True)
plt.subplots_adjust(wspace=0.1, hspace=0.1)

ax[0].plot(R2_list, c='k', ls='', marker='o', ms=5)

ax[1].plot(mse_test, ls='', marker='o', ms=5, label='test')
ax[1].plot(mse_train, ls='', marker='o', ms=5, label='train')

ax[2].plot(mae_test, ls='', marker='o', ms=5, label='test')
ax[2].plot(mae_train, ls='', marker='o', ms=5, label='train')

ax[0].set(ylabel='$R^2$', ylim=(0,1), yticks=np.arange(0,1.25,0.25))

ax[1].set(ylabel='MSE')
ax[-1].legend(loc=(0.02,1.15), ncol=1, scatterpoints=3)

ax[-1].set(ylabel='MAE', xticks=range(10), xlim=(-0.2,9.2))
ax[-1].set_xticklabels(xcol, rotation='vertical')

[axis.grid() for axis in ax];
````

### 3.2 Data preparation

Centering the data around zero and scaling the standard deviation to 1 is called standardization. If the features are on very different orders of magnite away from each other they might dominate the cost function and makes it very hard for the algorithm to converge.

### 3.3 Model prediction

Judging by the coefficient of determination which is $R^2<0.25$ for all attributes, we can say that none of the attributes can be reasonably fitted with a linear model of degree 1. Linear model is not appropriate especially for 3 attributes namely *low-density lipoproteins* and *high-density lipoproteins*, disease pregression doesn't seem to be dependent on these factors, and *sex*, which is just a binary label instead of a continuous variable and it's not meaningful to try to find a linear relation with disease progression.
