## Example for using color maps

```python
X = np.array(([-1,-1],[-1,1],[1,-1],[1,1]))
y = np.array((1,-1,-1,1))

cm_points = ListedColormap(["green", "blue"])
plt.figure(figsize=(7,7))

plt.scatter(X[:,0],X[:,1],marker='X',s=80,c=y,cmap=cm_points)
plt.xlabel('x1', fontsize=15)
plt.ylabel('x2', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.axis([-1.1,1.1,-1.1,1.1])
plt.show()
```

## Example using meshgrid

```python
h = 0.1
xx1, xx2 = np.meshgrid(np.arange(-1.1, 1.1, h), np.arange(-1.1, 1.1, h))
Y = np.sqrt(xx1**2 + xx2**2)

cm = plt.cm.Blues
fig,ax = plt.subplots(figsize=(5,5))

cs = ax.contourf(xx1, xx2, Y, cmap=cm, alpha=1.0)
ax.set_xlabel('x1')
ax.set_ylabel('x2')

plt.show()
```

## 1 Non-linear regression
### 1.1 Data acquisition

```python
np.random.seed(3)
```

```python
f = lambda x: x + 1.5*x**2 + 0.5*x**3 - 0.7*x**4 + np.sin(5*x) + np.cos(10*x)
```

```python
xall = np.linspace(-1, 2, 1000).reshape(-1, 1)
yall = f(xall).flatten()
```

```python
N = 40
x = 3*np.random.random((N,1)) - 1

rs = ShuffleSplit(n_splits=1, test_size=.25, random_state=0)

for train_idx, test_idx in rs.split(x):
    x_train, x_test = x[train_idx], x[test_idx]
    y_train, y_test = f(x_train).flatten(), f(x_test).flatten()
```

```python
fig, ax = plt.subplots()

ax.plot(xall, yall, c='k', lw=1.25, label='$f(x)$')
ax.scatter(x_train, y_train, s=15, label='training set')
ax.scatter(x_test, y_test, s=15, label='test set')

ax.set(xlabel='$x$', ylabel='$y$')
ax.legend(loc='best', scatterpoints=3);
```

### 1.2 – Support vector machine + linear kernel

- Definition of linear kernel

```python
svr_lin = SVR(kernel="linear", C=100)
svr_lin.fit(x_train, y_train)
```

```python
yall_pred = svr_lin.predict(xall)
y_train_pred = svr_lin.predict(x_train)  
y_test_pred = svr_lin.predict(x_test)
```

```python
fig, ax = plt.subplots()

ax.plot(xall, yall, c='k', lw=1.25, label='$f(x)$')
ax.plot(xall, yall_pred, c='C3', lw=1.25, label='prediction')
ax.scatter(x_train, y_train, s=15, label='training set')
ax.scatter(x_test, y_test, s=15, label='test set')

ax.set(xlabel='$x$', ylabel='$y$', title='SVM with linear kernel')
ax.legend(loc='best', scatterpoints=3);
```

```python
print('fitted slope: %g'%(m_fit:=svr_lin.coef_[0]))
print('fitted intercept: %g'%(b_fit:=svr_lin.intercept_))
```

```python
r2 = r2_score(y_test,y_test_pred)
print(f'R2 score = {r2:8.5f}')
    
mse_test = mean_squared_error(y_test,y_test_pred)
mse_train = mean_squared_error(y_train,y_train_pred)

mae_test = mean_absolute_error(y_test,y_test_pred)
mae_train = mean_absolute_error(y_test,y_test_pred)

print(f'MSE test  = {mse_test:8.5f}    MAE test  = {mae_test:8.5f}')
print(f'MSE train = {mse_train:8.5f}    MAE train = {mae_train:8.5f}') 
```

Looking at the plot an the $R^2$ score, we can tell that the function can not be well approximated using a linear kernel.

### 1.3  Support vector machine + Gaussian kernel

- Definition of Gaussian kernel

```python
GS = GridSearchCV(SVR(kernel='rbf'),
                   param_grid={'C':[1, 10, 100, 1000], 
                               'gamma':[1e-2, 1e-1, 1e0, 1e1, 1e2]},
                  scoring='neg_mean_squared_error', error_score=0)

GS.fit(x_train, y_train)

svr = SVR(kernel='rbf', **GS.best_params_)
svr.fit(x_train, y_train)
```

```python
yall_pred = svr.predict(xall)
y_train_pred = svr.predict(x_train)  
y_test_pred = svr.predict(x_test)
```

```python
fig, ax = plt.subplots()

ax.plot(xall, yall, c='k', lw=1.25, label='$f(x)$')
ax.plot(xall, yall_pred, c='C3', lw=1.25, label='prediction')
ax.scatter(x_train, y_train, s=15, label='training set')
ax.scatter(x_test, y_test, s=15, label='test set')

ax.set(xlabel='$x$', ylabel='$y$', title='SVM with Gaussian kernel')
ax.legend(loc='best', scatterpoints=3);
```

```python
r2 = r2_score(y_test,y_test_pred)
print(f'R2 score = {r2:8.5f}')
    
mse_test = mean_squared_error(y_test,y_test_pred)
mse_train = mean_squared_error(y_train,y_train_pred)

mae_test = mean_absolute_error(y_test,y_test_pred)
mae_train = mean_absolute_error(y_test,y_test_pred)

print(f'MSE test  = {mse_test:8.5f}    MAE test  = {mae_test:8.5f}')
print(f'MSE train = {mse_train:8.5f}    MAE train = {mae_train:8.5f}') 
```

Support vector regression with Gaussian kernel does an amazing job $(R^2 = 0.98)$ approximating this function.

### 1.4 Neural network

```python
n_nodes = np.arange(1,31,1)
MSE_train = np.zeros_like(n_nodes, dtype=float)
MSE_test = np.zeros_like(n_nodes, dtype=float)

for i in range(len(n_nodes)):

    nn = MLPRegressor(hidden_layer_sizes=(n_nodes[i]),
                      activation='logistic',
                      solver='lbfgs',
                      max_iter=10000)

    nn.fit(x_train, y_train)

    y_train_pred = nn.predict(x_train)  
    y_test_pred = nn.predict(x_test)
    
    MSE_test[i] = mean_squared_error(y_test,y_test_pred)
    MSE_train[i] = mean_squared_error(y_train,y_train_pred)
```

```python
fig, ax = plt.subplots()

ax.semilogy(n_nodes, MSE_train, lw=1.25, marker='o', ms=5, label='train')
ax.semilogy(n_nodes, MSE_test, lw=1.25, marker='o', ms=5, label='test')

ax.axvline(n_nodes[np.argmin(MSE_test)], lw=1.25, ls='--', c='C1',
           label='argmin MSE-test')

ax.grid()
ax.legend()

ax.set(xlabel='number of nodes', ylabel='MSE',
       title='single hidden layer');
```

```python
nn = MLPRegressor(hidden_layer_sizes=(n_nodes[np.argmin(MSE_test)]),
                  activation='logistic',
                  solver='lbfgs',
                  max_iter=10000)

nn.fit(x_train, y_train)
```

```python
yall_pred = nn.predict(xall)
y_train_pred = nn.predict(x_train)  
y_test_pred = nn.predict(x_test)
```

```python
fig, ax = plt.subplots()

ax.plot(xall, yall, c='k', lw=1.25, label='$f(x)$')
ax.plot(xall, yall_pred, c='C3', lw=1.25, label='prediction')
ax.scatter(x_train, y_train, s=15, label='training set')
ax.scatter(x_test, y_test, s=15, label='test set')

ax.set(xlabel='$x$', ylabel='$y$', title='NN with single hidden layer')
ax.legend(loc='best', scatterpoints=3);
```

```python
r2 = r2_score(y_test,y_test_pred)
print(f'R2 score = {r2:8.5f}')
    
mse_test = mean_squared_error(y_test,y_test_pred)
mse_train = mean_squared_error(y_train,y_train_pred)

mae_test = mean_absolute_error(y_test,y_test_pred)
mae_train = mean_absolute_error(y_test,y_test_pred)

print(f'MSE test  = {mse_test:8.5f}    MAE test  = {mae_test:8.5f}')
print(f'MSE train = {mse_train:8.5f}    MAE train = {mae_train:8.5f}')
```

```python
n_nodes = np.arange(1,16,1)
MSE_train = np.zeros_like(n_nodes, dtype=float)
MSE_test = np.zeros_like(n_nodes, dtype=float)

for i in range(len(n_nodes)):

    nn = MLPRegressor(hidden_layer_sizes=(n_nodes[i],n_nodes[i]),
                      activation='logistic',
                      solver='lbfgs',
                      max_iter=10000)

    nn.fit(x_train, y_train)

    y_train_pred = nn.predict(x_train)  
    y_test_pred = nn.predict(x_test)
    
    MSE_test[i] = mean_squared_error(y_test,y_test_pred)
    MSE_train[i] = mean_squared_error(y_train,y_train_pred)
```

```python
fig, ax = plt.subplots()

ax.semilogy(n_nodes, MSE_train, lw=1.25, marker='o', ms=5, label='train')
ax.semilogy(n_nodes, MSE_test, lw=1.25, marker='o', ms=5, label='test')

ax.axvline(n_nodes[np.argmin(MSE_test)], lw=1.25, ls='--', c='C1',
           label='argmin MSE-test')

ax.grid()
ax.legend()

ax.set(xlabel='number of nodes', ylabel='MSE',
       title='two hidden layers');
```

```python
nn = MLPRegressor(hidden_layer_sizes=(n_nodes[np.argmin(MSE_test)]),
                  activation='logistic',
                  solver='lbfgs',
                  max_iter=10000)

nn.fit(x_train, y_train)
```

```python
yall_pred = nn.predict(xall)
y_train_pred = nn.predict(x_train)  
y_test_pred = nn.predict(x_test)
```

```python
fig, ax = plt.subplots()

ax.plot(xall, yall, c='k', lw=1.25, label='$f(x)$')
ax.plot(xall, yall_pred, c='C3', lw=1.25, label='prediction')
ax.scatter(x_train, y_train, s=15, label='training set')
ax.scatter(x_test, y_test, s=15, label='test set')

ax.set(xlabel='$x$', ylabel='$y$', title='NN with two hidden layers')
ax.legend(loc='best', scatterpoints=3);
```

```python
r2 = r2_score(y_test,y_test_pred)
print(f'R2 score = {r2:8.5f}')
    
mse_test = mean_squared_error(y_test,y_test_pred)
mse_train = mean_squared_error(y_train,y_train_pred)

mae_test = mean_absolute_error(y_test,y_test_pred)
mae_train = mean_absolute_error(y_test,y_test_pred)

print(f'MSE test  = {mse_test:8.5f}    MAE test  = {mae_test:8.5f}')
print(f'MSE train = {mse_train:8.5f}    MAE train = {mae_train:8.5f}')
```
