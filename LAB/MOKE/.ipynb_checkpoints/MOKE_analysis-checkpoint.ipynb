```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

___
#### 1) Calibration

```python
cal_data = np.loadtxt('data/M11/calibration.txt')
```

```python
fig, ax = plt.subplots()

ax.plot(cal_data[:,0], cal_data[:,1], ls='', marker='.')
ax.set(xlabel='Current [A]', ylabel='Magnetic field [T]')
```

___
### 2) Hysteresis

```python
ω = np.arange(300, 375, 5)
data = [np.loadtxt('data/M11/hysteresis/'+str(ω_)) for ω_ in ω]
```

```python
fig, ax = plt.subplots(15, figsize=(6,25), dpi=100)

for i in range(len(ω)):
    #data[i][:,1] = data[i][:,1] - data[i][:,1].mean()
    #data[i][:,1] = data[i][:,1] / data[i][:,1].max()
    ax[i].plot(data[i][:,0], data[i][:,1], marker='.', ls='-')
    ax[i].set(ylabel=ω[i])
```

```python
fig, ax = plt.subplots()

for i in range(len(ω)-6):
    data[i][:,1] = data[i][:,1] - data[i][:,1].mean()
    data[i][:,1] = data[i][:,1] / data[i][:,1].max()
    ax.plot(data[i][:,0], data[i][:,1], marker='.', ls='-')
```
