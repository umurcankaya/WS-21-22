```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

#### 2.3. Fluorescence excitation and emission spectra

```python
λ_ex = np.arange(270,420+1,1)
A_ex = np.loadtxt('data/M11/A_excitation_270-420nm_1nm')

λ_em = np.arange(380,530+1,1)
A_em = np.loadtxt('data/M11/A_emission_380_530_1_A_Emisson.txt', skiprows=8)
```

```python
def model(λ, μ0, μ1, μ2, μ3, σ0, σ1, σ2, σ3, A0, A1, A2, A3, b):
    λ = λ.reshape((1,λ.shape[0]))
    μ = np.array((μ0, μ1, μ2, μ3)).reshape((4,1))
    σ = np.array((σ0, σ1, σ2, σ3)).reshape((4,1))
    A = np.array((A0, A1, A2, A3)).reshape((4,1))
    return np.sum(A * np.exp(-(λ-μ)**2/σ**2), axis=0) + b
```

```python
p0_ex = (380, 361, 345, 328, 10, 10, 5, 5, 0.8, 0.8, 0.5, 0.1, 1)
popt_ex, pcov = curve_fit(model, λ_ex, A_ex, p0_ex)

print('excitation')
for i in range(4):
    print(f'μ_{i} = {popt_ex[i] : .3f}')
```

```python
p0_em = (475, 450, 430, 407, 10, 10, 5, 5, 0.8, 0.8, 0.5, 0.1, 1)
popt_em, pcov = curve_fit(model, λ_em, A_em, p0_em)

print('emission')
for i in range(4):
    print(f'μ_{i} = {popt_em[i] : .3f}')
```

```python
fig, ax = plt.subplots()

ax.plot(λ_ex, A_ex, label='excitation')
ax.plot(λ_em, A_em, label='emision')

for i in range(4):
    ax.scatter(np.array([popt_ex[i]]), model(np.array([popt_ex[i]]), *popt_ex), marker='$'+str(i)+'$', c='k', s=50, zorder=10)
    ax.scatter(np.array([popt_em[i]]), model(np.array([popt_em[i]]), *popt_em), marker='$'+str(i)+'$', c='k', s=50, zorder=10)
    
ax.legend(loc='upper left')
ax.set(xlabel='Wavelenght [nm]', ylabel='Intensity [$\mu$A]', title='Sample A');
```

```python
Δλ = abs(popt_ex[1] - popt_em[2])
print(f'{Δλ = : .3f}')
```

#### 2.5 Time resolved fluorescence depolarization of DPH in DMPC vesicles

```python
fig, ax = plt.subplots()

ax.plot(np.loadtxt('data/irf.asc', skiprows=9, comments='*'))
ax.set(xlabel='$t$ [ns]', ylabel='count')
```

```python
temps = np.array((12.0, 23.3, 34.2))
data = {}

for temp in temps:
    data[temp] = np.loadtxt('data/%.1f.asc'%temp, skiprows=9, comments='*').reshape((4, 1024))
    
labels = ('HH', 'HV', 'VH', 'VV')
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12,3), sharey=True, sharex=True)

for i, (axis, temp) in enumerate(zip(ax, temps)):
    axis.plot(data[temp].T, label=labels)
    axis.set(title=f'$T = {temp}$'+'$\degree$C')
    
ax[-1].legend(title='Polarizer\nalignments')
    
plt.tight_layout()
```

```python
G = {}
R = {}
cutoff = 87
for temp in temps:
    I = data[temp]
    G[temp] = (I[1][cutoff:] / I[0][cutoff:]).mean()
    R[temp] = (I[3][cutoff:] - G[temp]*I[2][cutoff:]) / (I[3][cutoff:] + 2*G[temp]*I[2][cutoff:])
```

```python
fig, ax = plt.subplots()

for temp, c in zip(temps,['C4', 'C9', 'C6']):
    ax.plot(R[temp], label=temp, c=c, alpha=0.75)
    
ax.set(xlabel='$t$ [ns]', ylabel='r(t)')
ax.legend(title='Temperature [C]', loc='upper right')
```
