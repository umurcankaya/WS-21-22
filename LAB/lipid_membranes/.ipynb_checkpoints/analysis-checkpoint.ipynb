#### 2.5 Time resolved fluorescence depolarization of DPH in DMPC vesicles

```python
temps = np.array((12.0, 23.3, 34.2))
data = {}

for temp in temps:
    data[temp] = np.loadtxt('data/%.1f.asc'%temp, skiprows=9, comments='*').reshape((4, 1024))
```

```python
fig, ax = plt.subplots(1, 3, figsize=(14,4), sharey=True, sharex=True)

for axis, temp in zip(ax, temps):
    axis.plot(data[temp].T)
    axis.set(title=f'$T = {temp}$'+'$\degree$')
```
